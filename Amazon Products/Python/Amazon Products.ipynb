# Processamento de Dados

## Manipulação/Tratamento de Colunas Numéricas

```python
temp = ['discounted_price', 'actual_price', 'discount_percentage', 'rating_count']
for col in temp:
    df[col] = df[col].str.replace(r"[\₹,%]", "", regex=True).astype("float32")

df.head(1)
```

## Tratamento/Manipulação de Categorias

```python
df["category"] = df["category"].str.split("|")

df.head(1)
```

## Tratamento/Manipulação de Detalhes do Usuário

```python
temp = ["user_id", "user_name", "review_id", "review_title", "review_content"]
for col in temp:
    df[col] = df[col].str.split(",")

df.head(1)
```

## Tratamento de Problemas nas Avaliações (Ratings)

```python
df["rating"].max()
```

```python
df[df["rating"] == "|"] #Selecionar linhas problemáticas
```

```python
df.loc[df["rating"] == "|", "rating"] = "0" #Substituir o valor inválido por 0
```

```python
df["rating"] = df["rating"].astype("float16") #Converter para tipo numérico
```

# EDA

## Top 10 Produtos por Avaliação

```python
temp = df.sort_values("rating", ascending=False).head(10)

for i, (name, rating) in enumerate(zip(temp['product_name'], temp['rating'])):
    alpha = 1 - (i / len(temp))
    plt.barh(name[:30], rating, color='green', alpha=alpha)

plt.gca().invert_yaxis()
plt.xlabel('Avaliação')
plt.title('Top 10 Produtos por Avaliação')
plt.tight_layout()
plt.show()
```

## Top 10 Produtos por Quantidade de Avaliações

```python
temp = df.sort_values("rating_count", ascending=False).head(10)

for i, (name, rating) in enumerate(zip(temp["product_name"], temp["rating_count"])):
    alpha = 1 - (i / len(temp))
    plt.barh(name[:30], rating, color="green", alpha=alpha)

plt.gca().invert_yaxis()
plt.xlabel("Contagem de Avaliações")
plt.title("TTop 10 Produtos por Quantidade de Avaliações")
plt.tight_layout()
plt.show()
```

## Desconto vs Avaliação

```python
plt.scatter(df["discount_percentage"], df["rating"], color="green")
plt.xlabel("Discount")
plt.ylabel("Rating")
plt.title("Desconto vs Avaliação")
plt.tight_layout()
plt.show()
```

## Distribuição do Percentual de Desconto

```python
plt.figure(figsize=(10, 5))
plt.hist(df['discount_percentage'].dropna(), bins=30, color='green', edgecolor='black')
plt.title('Distribuição do Percentual de Desconto')
plt.xlabel('Percentual de Desconto')
plt.show()
```

## Distribuição das Avaliações dos Produtos

```python
plt.figure(figsize=(8, 5))
plt.hist(df['rating'], bins=20, color='lightgreen',edgecolor='black')
plt.title('Distribuição das Avaliações dos Produtos')
plt.xlabel('Avaliações')
plt.show()
```
